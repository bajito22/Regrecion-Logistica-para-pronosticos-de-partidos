```python
import requests
from bs4 import BeautifulSoup
full_links=[]
url = 'https://fbref.com/es/comps/9/2022-2023/horario/Marcadores-y-partidos-de-2022-2023-Premier-League'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Buscar todas las etiquetas <a> con el texto "Informe del partido"
etiquetas_a = soup.find_all('a', text='Informe del partido')

# Iterar sobre las etiquetas encontradas e imprimir los enlaces (href)
for etiqueta_a in etiquetas_a:
    enlace = etiqueta_a['href']
    etiqueta="https://fbref.com"+enlace
    full_links.append(etiqueta)
```

recorre una lista de URLs, realizando solicitudes, procesando los datos y almacenando la información correspondiente en diferentes listas. También parece haber un manejo adecuado de errores y tiempos de espera para evitar problemas con las solicitudes.

```python
Team1 = []
Team2 = []
TirosVisitante = []
TirosLocal = []
estatistics = []
formacionVisit = []
formacionLocal = []
```

```python
def verificationteam(TirosLocal_df,TirosVisitante_df):
    for i in range(len(TirosLocal_df)):
        booleanteam1=isinstance(TirosLocal_df["Equipo"].iloc[i],str)
        booleanteam2=isinstance( TirosVisitante_df["Equipo"].iloc[i],str)
        if booleanteam1:
            team1 = TirosLocal_df["Equipo"].iloc[i]    
        if booleanteam2:
            team2 = TirosVisitante_df["Equipo"].iloc[i]
        if booleanteam1 and booleanteam2:
            break    
    return team1,team2
```

```python
import pandas as pd
import time
from urllib.error import HTTPError

# Suponiendo que 'full_links' es tu lista de enlaces
for i in range(len(df_links)):
    # Lectura del link
    url = df_links["URLs"].iloc[i]
    try:
        print(f"Procesando URL {i + 1} de {len(df_links)}")

        # Realiza las solicitudes y extrae los datos
        read = pd.read_html(url)
        TirosLocal_df = pd.DataFrame(read[18])
        TirosVisitante_df = pd.DataFrame(read[19])

        try:
            # Intenta eliminar el primer nivel de las columnas si existe
            TirosVisitante_df.columns = TirosVisitante_df.columns.droplevel(level=0)
            TirosLocal_df.columns = TirosLocal_df.columns.droplevel(level=0)
        except ValueError:
            pass
        team1,team2=verificationteam(TirosLocal_df,TirosVisitante_df)
        # Equipo local y visitante desde datalimpia
        team1_1 = datalimpia["Local"].iloc[i]
        team2_1 = datalimpia["Visitante"].iloc[i]
        if (team1 == team1_1) and (team2 == team2_1):
            formacionLocal.append(read[0])
            formacionVisit.append(read[1])
            estatistics.append(read[2])
            Team1.append(team1)
            Team2.append(team2)
            TirosLocal.append(TirosLocal_df)
            TirosVisitante.append(TirosVisitante_df)
        else:
            print("---Algo salió mal---")
    except HTTPError as e:
        print(f"Error de solicitud al procesar la URL {url}: {e}")
    except Exception as e:
        print(f"Error al procesar la URL {url}: {e}")
        i=i-1
        time.sleep(5)  # Agrega un tiempo de espera antes de volver a intentar
    finally:
        time.sleep(10)  # Agrega un tiempo de espera entre solicitudes para evitar problemas con el servidor
```

```python
def organizarestatics(estatistics):
    """
    Extrae datos de las tablas estatics y los organiza
    Args:
    - estatistics (array): Datos posecion, precision, ...
    Returns:
    DataFrame : Lista con los datos organizados
    """
    data = []
    for i in range(len(estatistics)):
        pos1 = estatistics[i].iloc[0, 0]
        pos2 = estatistics[i].iloc[0, 1]
        pres1 = estatistics[i].iloc[2, 0]
        pres2 = estatistics[i].iloc[2, 1]
        dis1 = estatistics[i].iloc[4, 0]
        dis2 = estatistics[i].iloc[4, 1]
        Sal1 = estatistics[i].iloc[6, 0]
        Sal2 = estatistics[i].iloc[6, 1]
        data.append({
            "Pos.1":pos1,
            "Pos.2":pos2,
            "Pres.1":pres1,
            "Pres.2":pres2,
            "Dis.1":dis1,
            "Dis.2":dis2,
            "Sal.1":Sal1,
            "Sal.2":Sal2,
        })
    df=pd.DataFrame(data)
    return df    

    
```

```python
datos_organizados = organizarestatics(estatistics)
Pos_1=datos_organizados["Pos.1"]
Pos_2=datos_organizados["Pos.2"]
dfTeam1=pd.DataFrame(Team1)
dfTeam2=pd.DataFrame(Team2)
dfTeams = pd.concat([dfTeam1, df_Team2,Pos_1,Pos_2], axis=1)
dfTeams.columns = ["Local", "Visitante","Pos.1","Pos.2"]
```

```python
def itereitor2000(Tiros):
    dataframes = pd.DataFrame()

    for i in range(len(Tiros)):
        data = Tiros[i].copy()
        data = data.drop(["Notas"], axis=1)
        data = data.dropna(subset="Equipo",axis=0)
        # Agregar una columna con el iterador i
        data['Partido'] = (i+1)
    
        dataframes = pd.concat([dataframes, data], axis=0)
    return dataframes
```
